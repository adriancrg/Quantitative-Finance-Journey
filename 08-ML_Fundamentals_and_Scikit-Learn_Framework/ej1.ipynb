Enunciado: Siguiendo la estructura del ejemplo, construye un modelo de Machine Learning que intente predecir el precio Máximo del día (High) utilizando como características (features) el precio de Apertura (Open), el precio Mínimo (Low) y el Volumen (Volume) de ese mismo día.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = yf.download('MSFT', period='10y')

data['High_no_index'] = data['High']
# Better solution than mine: data.columns = data.columns.get_level_values(0)


X = data[['Open','Low','Volume']]
Y = data['High_no_index']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, shuffle=False)

print(f"Train set size: {len(X_train)} days")
print(f"Test set size: {len(X_test)} days")
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)

predictions = model.predict(X_test)

# Aplanamos el array de predicciones para que sea 1D
# predictions_flat = predictions.flatten()

"""
# Celda de depuración temporal
print("Shape de y_test:", Y_test.shape)
print("Shape de predictions:", predictions.shape)
print("\n")
print("Tipo de y_test:", type(Y_test))
print("Tipo de predictions:", type(predictions))

print(Y_test)

Y_test_1d = Y_test.ravel()
predictions_1d = predictions.ravel()
"""

results_df = pd.DataFrame({'Real': Y_test, 'Predicted': predictions})
print("\n--- First 5 predictions vs. real values ---")
print(results_df.head())
```

```python
print('Mean Absolut Error (MAE):', metrics.mean_absolute_error(Y_test, predictions))
```
