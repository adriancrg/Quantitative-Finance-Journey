```python
# Celda 1: Importar las librerías necesarias
# Damos "alias" a las librerías para escribir menos (pd, yf, plt). Es una convención estándar.
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

print("Librerías importadas correctamente.")
```

```python
# Celda 2: Descargar datos de NVDIA (NVDA)
# Definimos el 'ticker' (identificador del activo) y el periodo.
ticker = "NVDA"
# Usamos yfinance para descargar los datos. Devuelve un DataFrame de pandas.
nvda_data = yf.download(ticker, start="2015-01-01", end="2025-10-16", auto_adjust=True)
```

```python
# Celda 3: Calcular una media móvil rápida de 20 días y otra lenta de 100 días

# Creamos una nueva columna en nuestro DataFrame para almacenar el resultado.
nvda_data['MA20'] = nvda_data['Close'].rolling(window=20).mean()

# Creamos una nueva columna en nuestro DataFrame para almacenar el resultado.
nvda_data['MA100'] = nvda_data['Close'].rolling(window=100).mean()
```

```python
# Celda 4: Visualizar el precio de cierre y la media móvil
# plt.figure() nos permite definir el tamaño del gráfico.
plt.figure(figsize=(14, 7))

# Graficamos la columna 'Close'
plt.plot(nvda_data['Close'], label='Precio de Cierre NVDA')

# Graficamos la columna 'MA20'
plt.plot(nvda_data['MA20'], label='Media Móvil 20 días', color='orange')

# Graficamos la columna 'MA100'
plt.plot(nvda_data['MA100'], label='Media Móvil 100 días', color='red')

# Añadimos contexto al gráfico para que sea legible y profesional.
plt.title(f'Precio de Cierre y Medias Móviles de 20 y 100 días para {ticker}')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.legend()  # Muestra las etiquetas (labels) que definimos en los plots.
plt.grid(True) # Añade una cuadrícula para facilitar la lectura.
plt.show() # Muestra el gráfico.
```
