```python
#Continuación ejercicio 1 módulo 8:

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = yf.download('MSFT', period='10y')
data.columns = data.columns.get_level_values(0)

X = data[['Open','Low','Volume']]
y = data['High']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)

print(f"Train set size: {len(X_train)} days")
print(f"Test set size: {len(X_test)} days")

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

"""
# Celda de depuración temporal
print("Shape de y_test:", y_test.shape)
print("Shape de predictions:", predictions.shape)
print("\n")
print("Tipo de y_test:", type(y_test))
print("Tipo de predictions:", type(predictions))

print(y_test)

y_test_1d = y_test.ravel()
predictions_1d = predictions.ravel()
"""

results_df = pd.DataFrame({'Real': y_test, 'Predicted': predictions})
print("\n--- First 5 predictions vs. real values ---")
print(results_df.head())

print('Mean Absolut Error (MAE):', metrics.mean_absolute_error(y_test, predictions))
```

```python
#1
print("--- Interpretación del Modelo Entrenado ---")
print(f"Coeficientes (w_1, w_2, w_3): {model.coef_}")
print("w_1 nos dice cúanto subirá el High por cada dólar que suba el Open, el w_2 en función del Low y el w_3 en función del Volumen")
print(f"Intercept (w_0): {model.intercept_:.4f}")
print("El intercept (w_0) es el valor base")
print("\n")
```

```python
#2

print("--- Evaluación del Modelo de ML ---")
mae = metrics.mean_absolute_error(y_test, predictions)
rmse = np.sqrt(metrics.mean_squared_error(y_test, predictions))
r2 = metrics.r2_score(y_test, predictions)

print(f"Error Absoluto Medio (MAE): {mae:.2f} dólares")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse:.2f} dólares")
print(f"R-cuadrado (R²): {r2:.2%}") # Formateado como porcentaje
print("\n")
```

```python
#3,4 y 5

print("--- Evaluación del Modelo Baseline (Predecir el precio de hoy) ---")
print("Usamos como modelo baseline que el precio High del día sea el mismo que el Open")
baseline_predictions = X_test['Open']

baseline_mae = metrics.mean_absolute_error(y_test, baseline_predictions)
baseline_rmse = np.sqrt(metrics.mean_squared_error(y_test, baseline_predictions))
baseline_r2 = metrics.r2_score(y_test, baseline_predictions)

print(f"Baseline MAE: {baseline_mae:.2f} dólares")
print(f"Baseline RMSE: {baseline_rmse:.2f} dólares")
print(f"Baseline R²: {baseline_r2:.2%}")
print("\n")

# --- 6. Conclusión ---
print("--- Conclusión ---")
if rmse < baseline_rmse:
    print("¡Éxito! Nuestro modelo de ML es ligeramente mejor que la baseline.")
else:
    print("Nuestro modelo de ML no supera a la simple estrategia de la baseline.")
```

```python
#mejor interpretación de pesos


# --- 1. Interpretación del Modelo Entrenado ---

# Creamos una Serie de Pandas para ver los coeficientes con sus nombres. ¡Mucho más claro!
coeficientes = pd.Series(model.coef_, index=X_train.columns)

print("--- Interpretación del Modelo Entrenado ---")
print("Coeficientes:")
print(coeficientes)
print(f"\nIntercepto (w_0): {model.intercept_:.4f}")

print("\n--- Análisis de los Coeficientes ---")
print("1. Open: Por cada $1 que aumenta el precio de apertura, el modelo predice que el 'High' aumentará en ~$0.99, manteniendo lo demás constante.")
print("2. Low: Por cada $1 que aumenta el precio mínimo, el modelo predice que el 'High' aumentará en ~$0.20.")
print("3. Conclusión de Impacto: Los coeficientes de 'Open' y 'Low' son los más significativos. Un valor cercano a 1 para 'Open' es muy intuitivo. El coeficiente de 'Volume' es prácticamente cero, lo que sugiere que en su escala actual, no tiene un impacto lineal directo en la predicción del 'High'.")

```

```python
# Mejora de conclusión
print("--- Conclusión Final: Modelo ML vs. Baseline ---")

print(f"El modelo de Machine Learning (RMSE: {rmse:.2f}) es significativamente más preciso que el modelo baseline (RMSE: {baseline_rmse:.2f}).")
print(f"El modelo de ML reduce el error de predicción en aproximadamente un {( (baseline_rmse - rmse) / baseline_rmse ):.1%} en comparación con la simple estrategia de usar el precio de apertura.")
print(f"Además, el R² de {r2:.2%} indica que nuestro modelo explica el {r2:.2%} de la variabilidad en el precio 'High', mientras que la baseline solo explica el {baseline_r2:.2%}.")
print("\nConclusión: Sí, el modelo de Machine Learning es sustancialmente mejor y aporta un valor predictivo real.")
```
